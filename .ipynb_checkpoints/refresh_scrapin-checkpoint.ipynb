```python
from bs4 import BeautifulSoup
from urllib import request
import re
import csv
```

```python
html = request.urlopen("http://localhost:39233/")
bs = BeautifulSoup(html, "lxml")
a = bs.find_all("a", href = re.compile(r'(\.html)$'))
```

```python
with open("locals.csv", "w+", encoding="utf-8") as f:
    for anchor in a:
        title, href = anchor.text.strip('_ '), anchor.attrs.get('href')
        print("Written: ", end="")
        print(anchor.text.strip('_ '), anchor.attrs.get('href'))
        writer = csv.writer(f)
        writer.writerow([title, ": ", href])
    f.close()
```
